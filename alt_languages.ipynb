```python
# write down all needed imports
import os
import pandas as pd
from pathlib import Path
import json
```

```python

 #designate the csv file data path
data_load = Path ('Resources/data.csv')

# uniform language
language = pd.read_csv(data_load,encoding="utf-8")
language.head()
```

```python
# create dataframe that drops unnecessary columns
language_df = language.drop(["ID", "Name in French", 'Name in Spanish', 'Country codes alpha 3', 
                                      'Name in the language', 'Sources', 'Description of the location' ], axis=1)

# drop duplicates, fill NaN object values with "None"
language_df= language_df.drop_duplicates()
language_df = language_df.dropna()
language_df
```

```python
language_df.dtypes
```

```python
alt_language_df = language_df[['Countries', 'Latitude', 'Longitude']]
# drop duplicates, fill NaN object values with "None"
alt_language_df.drop_duplicates()
alt_language_df.dropna()
alt_language_df
```

```python

geojson = {"type": "FeatureCollection", "features": []}

for _, row in alt_language_df.iterrows():
    feature = {"type": "Feature", "geometry": {"type": "Point", "coordinates": [row['Longitude'], row['Latitude']]}, "properties": {"country": row['Countries']}}
    geojson['features'].append(feature)

with open('result.geojson', 'w') as fp:
    json.dump(geojson, fp)  
```
