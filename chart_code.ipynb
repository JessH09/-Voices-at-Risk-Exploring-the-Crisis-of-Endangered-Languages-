```python
# pie chart
endangerment_counts = df['Degree of endangerment'].value_counts()

plt.figure(figsize=(12, 8))
plt.pie(endangerment_counts, labels=endangerment_counts.index, autopct='%1.1f%%', shadow=False, startangle=140, textprops={'fontsize': 10, 'fontweight': 'bold'})
plt.title('Distribution of Endangered Languages', fontsize=12)


plt.savefig('pie_chart.png', dpi=400)

plt.show()
```
