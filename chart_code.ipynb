```python
# pie chart
endangerment_counts = df['Degree of endangerment'].value_counts()

plt.figure(figsize=(8, 5))
plt.pie(endangerment_counts, labels=endangerment_counts.index, autopct='%1.1f%%', shadow=False, startangle=140, textprops={'fontsize': 8, 'fontweight': 'bold'})
plt.title('Distribution of Endangered Languages', fontsize=12)


plt.savefig('pie_chart_update.png',bbox_inches='tight')

plt.show()
```
